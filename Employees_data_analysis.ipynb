```python
import pandas as pd
df = pd.read_csv('employees.csv',parse_dates=['Start Date','Last Login Time'])
df.head(10)
```

```python
df['Senior Management']=df['Senior Management'].astype('bool')
df['Gender']= df['Gender'].astype('category')
df.info()
```

```python
# ONly male
df[df['Gender']== 'Male']
```

```python
# Team = Finance
df[df['Team']== 'Finance']
```

```python
# Senior mgt is True
df[df['Senior Management']== True]
```

```python
# NOt in the team mkting 
df[df['Team']!= 'Marketing']
```

```python
# salary >110000
df[df['Salary']> 110000]
```

```python
# BOnus <1.15
df[df['Bonus %']< 1.15]
```

```python
# Started before Jan 1,1985
df[df['Start Date']< '1985-01-01']
```

```python
# (Team = mkting and gender = Female) or (Gender = male and senior management = true)

filter1= (df['Team']== 'Marketing') & (df['Gender']== 'Female')
filter2= (df['Senior Management']== True) & (df['Gender']== 'Female')
df[filter1|filter2]
```

```python
 # team = marketing, sales or product
df[df['Team'].isin(['Marketing','Sales','Product'])]
```

```python
# salary between 60000 and 80000
df[df['Salary'].between(60000,80000)]
```

```python
# total days spent in the organisation from start date , rounded upto 2 decimal points
# finding out top 5 employees who spent the most time in the organisation
today=pd.Timestamp.today().normalize()
df['Days spent']=(today-df['Start Date'])
df['Years spent']=((today-df['Start Date']).dt.days/365).round(2)
df['Years spent']
df.sort_values(by=['Years spent'],ascending=False).head(5)
```

```python
# most hours spent in the org
today
total_time = today + pd.Timedelta(hours=23, minutes=59)
total_time
df['Last Login Time']
df['Time spent']= (total_time - df['Last Login Time'])
df['Hours spent'] = (df['Time spent'].dt.total_seconds()/3600).round(2)
df
```

# FUNCTIONAL METHOD OF CREATING MATPLOTLIB
## (plt.plot and plt.subplot)

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 


x = np.linspace(0,5,11)
y = x**2

plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Basic chart title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'red')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

# OBJECT ORIENTED METHOD OF MATPLOTLIB
## Create figure objects and then call methods off of that
### (figure, axes , axes.plot)

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)

axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title of the chart')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)

axes2.set_title('Axes2 plot title here')
axes1.set_title('Axes1 plot title here')
```

# Subplots

### Looping through axes 

```python
fig,axes = plt.subplots(1,2)

for i in axes:
    i.plot(x,y)

plt.tight_layout()
```

### Manual axes index

```python
fig,axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[0].set_title('First plot')

axes[1].plot(y,x)
axes[1].set_title('Second plot')

plt.tight_layout()
```

# Figure size, aspect ratio & DPI 

### For normal plots

```python
fig = plt.figure(figsize=(5,3))

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y)
```

### For SUBplots

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(5,3))

axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

# ADding legend in plots (2 step process)
## *first step : include label in .plot()
## *second step : call .legend() method

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y,label ='x,y line')
axes.plot(y,x,label ='y,x line')
axes.plot(x**2,label ='X squared line')
axes.plot(x**3,label ='X cubed line')

axes.set_title('Multi line in 1 plot')
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.legend(loc=0)
```

# line COLOR='red', LineWIDTH=2 , TRansparency(alpha=0.5)

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y,label ='x,y line',linewidth=4,color='red',alpha=.5)
axes.plot(y,x,label ='y,x line',linewidth=4,color='blue',alpha=.5)
axes.plot(x**2,label ='X squared line',linewidth=4,color='#D3AF37',alpha=.5)
axes.plot(x**3,label ='X cubed line',linewidth=4,color='#eea9a9',alpha=.5)

axes.set_title('Multi line in 1 plot')
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.legend(loc=0)
```

# LINE STYLE '--' ,  ':' ,  '-.'  , 'solid' ,   'dashed',   'dashdot',  'dotted'

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y,label ='x,y line',linewidth=3,linestyle='--')
axes.plot(y,x,label ='y,x line',linewidth=3,linestyle='-.')
axes.plot(x**2,label ='X squared line',linewidth=3,linestyle='dotted')
axes.plot(x**3,label ='X cubed line',linewidth=3,linestyle='-')

axes.set_title('Multi line in 1 plot')
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.legend(loc=0)
```

# Marker size and market style

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y,label ='x,y line',linewidth=3,linestyle='--',marker='o',markersize=8,markerfacecolor='yellow')
axes.plot(y,x,label ='y,x line',linewidth=3,linestyle='-.',marker='1',markersize=8)
axes.plot(x**2,label ='X squared line',linewidth=3,linestyle='dotted',marker='H',markersize=8)
axes.plot(x**3,label ='X cubed line',linewidth=3,linestyle='-',marker='*',markersize=8)

axes.set_title('Multi line in 1 plot')
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.legend(loc=0)
```
